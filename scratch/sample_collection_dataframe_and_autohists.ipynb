```python
# Create the initial DataFrame template with blank dates
data_template = {
    "time_collected": [None]*11,
    "biomass_type": ["S"]*11,
    "processing_type": ["GD"]*11,
    "batch_number": [1]*11,
    "subsample_id": [1]*11,
    "measurement_type": ["EA"]*1 + ["SAC"]*5 + ["SAP"]*5,
    "run_id": [1]*1 + list(range(1, 6))*2,
    "carbon_content": [None]*11,
    "carbon_content_uncertainty": [None]*11,
    "nitrogen_content": [None]*11,
    "nitrogen_content_uncertainty": [None]*11,
    "sample_id": [None]*11
}
df_template = pd.DataFrame(data_template)

# Define the additional subsample_ids, shifts, and biomass_types
subsample_ids = list(range(1, 11))
batch_numbers = list(range(1,7))
biomass_types = ["H", "R", "S"]

# Function to generate sample_id
def generate_sample_id(row):
    return f"{row['biomass_type']}_{row['processing_type']}_{row['batch_number']}_{row['subsample_id']}_{row['measurement_type']}_{row['run_id']}"

# Create a list to hold all the DataFrames
dfs = []

# Loop through the hierarchical values and generate DataFrames
for biomass_type in biomass_types:
    for batch_number in batch_numbers:
        for subsample_id in subsample_ids:
            # Copy the template DataFrame
            new_df = df_template.copy()
            # Update the fields with current loop values
            new_df['biomass_type'] = biomass_type
            new_df['batch_number'] = batch_number
            new_df['subsample_id'] = subsample_id
            new_df['sample_id'] = new_df.apply(generate_sample_id, axis=1)
            dfs.append(new_df)

# Concatenate all DataFrames
df1 = pd.concat(dfs, ignore_index=True)
```

```python
def combined_plots(df, output_pdf):
    """
    Take in a dataframe with carbon content and generate side-by-side plots for each biomass type.
    On the left: Histogram with KDE for Carbon Content.
    In the middle: Variance vs. Sample Size for Carbon Content.
    On the right: Histogram with KDE for Moisture Content at Facility.
    Show the plots in the notebook and save them to a PDF file.
    """
    
    def calculate_variances(content_col, subset_df):
        variances = []
        sample_sizes = []
        valid_df = subset_df.dropna(subset=[content_col])
        for i in range(2, len(valid_df) + 1):
            sample = valid_df[content_col].iloc[:i]
            variance = np.var(sample)
            variances.append(variance)
            sample_sizes.append(i)
        return sample_sizes, variances
    
    plt_df = df[df.measurement_type == 'EA']
    biomass_types = plt_df['biomass_type'].unique()

    with PdfPages(output_pdf) as pdf:
        for biomass_type in biomass_types:
            subset_df = plt_df[plt_df['biomass_type'] == biomass_type]

            # Calculate variances for carbon_content
            sample_sizes_c, variances_c = calculate_variances('carbon_content', subset_df)

            # Create the figure with three subplots
            fig, ax = plt.subplots(1, 3, figsize=(21, 5))

            # Plot for carbon content histogram and KDE
            sns.histplot(subset_df['carbon_content'].dropna(), bins=25, kde=True, ax=ax[0])
            ax[0].set_title(f'Carbon Content Histogram with KDE for {biomass_type}')
            ax[0].set_xlabel('Carbon Content')
            ax[0].set_ylabel('Frequency')

            # Plot for variance vs. sample size
            sns.lineplot(x=sample_sizes_c, y=variances_c, marker='o', ax=ax[1])
            ax[1].set_title(f'Variance of Carbon Content vs. Sample Size for {biomass_type}')
            ax[1].set_xlabel('Sample Size')
            ax[1].set_ylabel('Variance')

            # Plot for moisture content at facility histogram and KDE
            sns.histplot(subset_df['moisture_content_at_facility'].dropna(), bins=25, kde=True, ax=ax[2])
            ax[2].set_title(f'Moisture Content at Facility Histogram with KDE for {biomass_type}')
            ax[2].set_xlabel('Moisture Content at Facility')
            ax[2].set_ylabel('Frequency')

            plt.tight_layout()
            
            # Save the current figure to the PDF
            pdf.savefig(fig)
            
            # Show the plot in the notebook
            plt.show()

            # Close the figure to avoid overlap
            plt.close()
```
