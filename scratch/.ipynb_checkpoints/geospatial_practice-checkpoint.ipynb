```python
# Shuffle the rows of the DataFrame randomly
df_shuffled = groceries.sample(frac=1, random_state=42)  # Use a random state for reproducibility

# Calculate the index to split the DataFrame in half
split_index = len(df_shuffled) // 2

# Split the DataFrame into two halves
df_half1 = df_shuffled.iloc[:split_index]
df_half2 = df_shuffled.iloc[split_index:]
```

```python
p1 = df_half1.iloc[0].geometry

l = [Point(p.x, p.y) for p in p1.geoms]
```

```python
fig, ax = plt.subplots()

df_half1.plot(ax=ax, color='red')
df_half2.plot(ax=ax, color='blue')

ax.scatter(l[0].x, l[0].y, color = 'black')
```

```python
gf = gpd.sjoin_nearest(df_half1, df_half2, distance_col="distances")
```

```python
gf
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from itertools import cycle

# Assuming `gf` is your DataFrame with columns Ycoord_left, Xcoord_left, Ycoord_right, Xcoord_right

# Create a new figure
plt.figure()

# Define a marker cycle
markers = cycle(['o', 's', 'D', '^', 'v', '>', '<', 'P', '*', 'X', 'H'])

# Iterate over each row of the DataFrame
for idx, row in gf.iterrows():
    # Extract coordinates for left and right points
    left_coords = (row['Xcoord_left'], row['Ycoord_left'])
    right_coords = (row['Xcoord_right'], row['Ycoord_right'])
    
    # Get the next marker type from the cycle
    marker = next(markers)
    
    # Plot left and right points with a different marker for each row
    plt.plot(*left_coords, marker, color='blue')  
    plt.plot(*right_coords, marker, color='red')  

# Add labels and title
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.title('Coordinates from DataFrame')

# Show plot
plt.show()
```
