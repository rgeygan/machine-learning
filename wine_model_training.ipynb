```python
import pandas as pd
import numpy as np
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import f1_score
from sklearn.metrics import mean_absolute_error

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = [
    "fixed acidity", "volatile acidity", "citric acid", 
    "residual sugar", "chlorides", "free sulfur dioxide", 
    "total sulfur dioxide", "density", "pH", "sulphates", 
    "alcohol", "quality"
]

# Read the CSV file with specified column names
df = pd.read_csv('/Users/ryangeygan/hacking/Datasets/winequality-red.csv', names=column_names)
```

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
X = df.drop(columns='quality')

y = df[['quality']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=41)

#XBG classifer needs classes to start at 0
le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
model = xgb.XGBClassifier(
                          objective='binary:logistic',
                            eval_metric='aucpr', #use AUC to evaluate classification
                            early_stopping_rounds=10, #when the prediction stops improving significantly, 10 more trees will be built and if those don't improve the preds, the algorithm stops
                            seed = 67)
```

```python
trained_model = model.fit(X_train, y_train, verbose=False, eval_set=[(X_test, y_test)])
```

```python
predictions = trained_model.predict(X_test)

backtransformed_preds = le.inverse_transform(predictions)
backtransformed_test = le.inverse_transform(y_test)
```

```python
# Get unique labels
unique_labels = le.inverse_transform(trained_model.classes_)

# Calculate confusion matrix
cm = confusion_matrix(backtransformed_test, backtransformed_preds, labels=unique_labels)

# Plot confusion matrix with specified labels
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_labels)
disp.plot()

plt.show()
```

```python
f1 = f1_score(backtransformed_test, backtransformed_preds, average='weighted')
mae = mean_absolute_error(backtransformed_test, backtransformed_preds)
```

```python
print(f"f1 accuracy is {f1.round(2)} and mean absolute error is {mae}")
```

```python
# Initialize counter
count_within_range = 0

# Iterate over elements of the arrays
for val1, val2 in zip(backtransformed_test, backtransformed_preds):
    if abs(val1 - val2) <= 1:
        count_within_range += 1

# Calculate percentage
percentage_within_range = (count_within_range / len(backtransformed_test)) * 100

print(f"Percentage of predictions within 1 quality point {percentage_within_range}%")
```
