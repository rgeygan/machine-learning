```python
# Path to the TIFF file
raster_file = '/Users/ryangeygan/hacking/Datasets/no2_image.tiff'

with rasterio.open(raster_file) as src:
    raster_data = src.read(1)  # Assuming it's a single band image, adjust the index accordingly

    # Validate CRS
    if not counties.crs == src.crs:
        raise ValueError("CRS of GeoDataFrame and raster do not match")

    # Iterate over each polygon in the GeoDataFrame
    computed_values = []
    for idx, county in counties.iterrows():
        mask = rasterio.features.geometry_mask([county.geometry], out_shape=(src.height, src.width), transform=src.transform, invert=True)
        values = raster_data[mask]

        computed_values.append(np.percentile(values, 80))

counties['80p_no2'] = computed_values
```

## Is there Global (overall) Spatial Autocorrelation (SA) in Atmospheric NO2 concentrations in the US?

```python
# Calculate spatial weights matrix

w = weights.distance.KNN.from_dataframe(counties, k=4)

w.transform = 'r'
```

```python
#Center the data of interest

counties['centered_data'] = counties['80p_no2'] - counties['80p_no2'].mean()
```

```python
counties['no2_lag'] = weights.lag_spatial(w, counties.centered_data)
```

```python
counties.head()
```

#### Looks like we have clustered data (as opposed to perfect separation or randomness)

```python
f, ax = plt.subplots(1, figsize=(6, 6))
sns.regplot(
    x="centered_data",
    y="no2_lag",
    ci=None,
    data=counties,
    line_kws={"color": "r"},
)
ax.axvline(0, c="k", alpha=0.5)
ax.axhline(0, c="k", alpha=0.5)
ax.set_title("Moran Plot - % Leave")
plt.show()
```

### There is a strong relationship between the Mean value of a neighborhood and the value of the center of the neighborhood

This means that as the concentration of no2 increases for a given county, the concentration of no2 of neighboring counties also tends to increase and vice versa for decreasing concentrations. In other words, there is evidence here that no2 concentrations are spatially clustered

```python
#Moran's I is similar to a correlation coefficient but in this case measuring how well the covariance of neighborhoods explains the overall variance

moran = esda.moran.Moran(counties["80p_no2"], w)

moran.I
```

## Is there Local SA - Are there hotspots, coldspots, or islands?

```python
lisa = esda.moran.Moran_Local(counties["80p_no2"], w)
```

```python
# Set up figure and axes
f, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
# Make the axes accessible with single indexing
axs = axs.flatten()

# Subplot 1 #
# Choropleth of local statistics
# Grab first axis in the figure
ax = axs[0]
# Assign new column with local statistics on-the-fly
counties.assign(
    Is=lisa.Is
    # Plot choropleth of local statistics
).plot(
    column="Is",
    cmap="plasma",
    scheme="quantiles",
    k=5,
    edgecolor="white",
    linewidth=0.1,
    alpha=0.75,
    legend=True,
    ax=ax,
)

# Subplot 2 #
# Quadrant categories
# Grab second axis of local statistics
ax = axs[1]
# Plot Quadrant colors (note to ensure all polygons are assigned a
# quadrant, we "trick" the function by setting significance level to
# 1 so all observations are treated as "significant" and thus assigned
# a quadrant color
esdaplot.lisa_cluster(lisa, counties, p=1, ax=ax)

# Subplot 3 #
# Significance map
# Grab third axis of local statistics
ax = axs[2]
#
# Find out significant observations
labels = pd.Series(
    1 * (lisa.p_sim < 0.05),  # Assign 1 if significant, 0 otherwise
    index=counties.index  # Use the index in the original data
    # Recode 1 to "Significant and 0 to "Non-significant"
).map({1: "Significant", 0: "Non-Significant"})
# Assign labels to `db` on the fly
counties.assign(
    cl=labels
    # Plot choropleth of (non-)significant areas
).plot(
    column="cl",
    categorical=True,
    k=2,
    cmap="Paired",
    linewidth=0.1,
    edgecolor="white",
    legend=True,
    ax=ax,
)


# Subplot 4 #
# Cluster map
# Grab second axis of local statistics
ax = axs[3]
# Plot Quadrant colors In this case, we use a 5% significance
# level to select polygons as part of statistically significant
# clusters
esdaplot.lisa_cluster(lisa, counties, p=0.1, ax=ax)

# Figure styling #
# Set title to each subplot
for i, ax in enumerate(axs.flatten()):
    ax.set_axis_off()
    ax.set_title(
        [
            "Local Statistics",
            "Scatterplot Quadrant",
            "Statistical Significance",
            "Moran Cluster Map",
        ][i],
        y=0,
    )
# Tight layout to minimize in-between white space
f.tight_layout()

# Display the figure
plt.show()
```
