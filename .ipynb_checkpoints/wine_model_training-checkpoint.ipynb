```python
import dill

import pandas as pd
import numpy as np
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import f1_score
from sklearn.metrics import mean_absolute_error

import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Labeled Feature Dataset

```python
column_names = [
    "fixed acidity", "volatile acidity", "citric acid", 
    "residual sugar", "chlorides", "free sulfur dioxide", 
    "total sulfur dioxide", "density", "pH", "sulphates", 
    "alcohol", "quality"
]

# Read the CSV file with specified column names
df = pd.read_csv('/Users/ryangeygan/hacking/Datasets/winequality-red.csv', names=column_names)
```

### Show Data Types and Ranges of Values

We'll use this later in the frontend web app and the backend data validation

```python
print("Column Information:")
for column in df.columns:
    col_type = df[column].dtype
    col_range = f"{df[column].min()} - {df[column].max()}"
    print(f"{column}: {col_type}, Range: {col_range}")
```

### Get a Quick Sense of Strength of Predictors

In this case we would actually want to use spearman rank or some other approach to evaluate the strength of the relationship between the features and the target because the target are labels, not a continuous variable

```python
# Calculate correlation matrix
corr_matrix = df.corr()

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### Construct an Estimator and Train the Model

```python
model = xgb.XGBClassifier(
                          objective='binary:logistic',
                            eval_metric='aucpr', #use AUC to evaluate classification
                            early_stopping_rounds=10, #when the prediction stops improving significantly, 10 more trees will be built and if those don't improve the preds, the algorithm stops
                            seed = 67)
```

```python
X = df.drop(columns='quality')

y = df[['quality']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=41)

#XBG classifer needs classes to start at 0
le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
trained_model = model.fit(X_train, y_train, verbose=False, eval_set=[(X_test, y_test)])
```

### Evaluate the Model Performace

```python
predictions = trained_model.predict(X_test)

backtransformed_preds = le.inverse_transform(predictions)
backtransformed_test = le.inverse_transform(y_test)
```

```python
# Get unique labels
unique_labels = le.inverse_transform(trained_model.classes_)

# Calculate confusion matrix
cm = confusion_matrix(backtransformed_test, backtransformed_preds, labels=unique_labels)

# Plot confusion matrix with specified labels
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=unique_labels)
disp.plot()

plt.show()
```

```python
f1 = f1_score(backtransformed_test, backtransformed_preds, average='weighted')
mae = mean_absolute_error(backtransformed_test, backtransformed_preds)

print(f"f1 score is {f1.round(2) * 100}% and mean absolute error is {mae.round(2)}")
```
