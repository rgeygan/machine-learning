## Download / Import Data and Inspect

1. Look at column names
2. Checking data types in columns
3. Check index
4. Check the size of the data
5. Identify targets and responses

## Check for Missing Data and Gap Filling

1. Check how much data is missing from each column
2. Options for dealing with missingness
    a. Some modeling algorithms can deal with missingness so it might be ok to leave as is
    b. Drop all rows with missing data (if missingness is relatively low) - but think about how to deal with a new feature vector that is missing data
    c. Interpolation - there are multiple methods to do this

## One-hot encoding / Dummy Vars for Categorical Data (discrete data)

1. There is an important note here about not introducing bias - encode data in a way that doesn't introduce correlations for string data
2. When there are only a few categories we can use dummy vars (0 / 1) encoding but if there are tons we don't want to end up with tons of extra columns
3. There are many options - one hot (only a few discrete values), label (when there are many values and we use means of targets as labels), and target encoding (when we want to use label encoding but the discrete vars are imbalanced or one class has relatively few examples)

## Other Data Preprocessing

1. Changing data types if necessary - temporal data types, strings, floats vs. int, etc.
2. Centering features - mostly if you are concerned about interpreting magnitude of coefficients as relative effects
3. Transformations of feature data - log transforms, eg. - this is important to satisfy assumptions for parametric models and would usually come after EDA

## EDA

1. Here is where we would inspect the feature vector for balance
    a. are any feature vectors distributed in strange ways that could affect the model?
    b. if multiple treatment groups can use SMDs
3. Look at correlations in the data
    a. looking for multicollinearity issues
    b. getting a sense of which vars have strongest correlation with target
4. Looking for normality if building a parametric model (qq plots or statsmodels.stats.diagnostic.kstest_normal())

## Feature Engineering

1. Create new features from extant data
    a. Eg. - creating an hour, day, month column for time series data
    b. Creating other features from seasonal data like peak values, inflection point dates
    c. Creating interaction columns - multiplying two values together to create a new interaction feature

## Train / Test Splitting and batching for Cross Validation

1. Be cautious - may want to use stratification to do train-test splitting if some features are highly imbalanced

## Building and Fitting a Model

1. Potentially building and fitting multiple models if we want to compare
2. Here we could also build a dummy classifier to test against a model that is randomly guessing

## Validating

1. Depending on the model, choosing an evaluation criterion ahead of time
    a. Parametric
       i. Regression - p-values, confidence intervals, rmse, aic, bic, and adjusted r^2
       ii. Classification (logistic regression) - psuedo r^2, class accuracy (like f1)
    b. Non-parametric
       i. Regression - calc RMSE, MAE, MAPE for test data and perhaps refitting and revalidating if doing cross validation (this is how we get a sense of model variance and bias)
       ii. Classification - precision, recall, F1, roc, auc
       iii. Recommender - precision and recall at K
       iv. Time series forecasting - confidence intervals
       v. A/B testing - parametric inference
       vi. Bayesian world - standard devitions of various normals that capture uncertainty of beliefs

## Hyper Parameter Tuning / Optimization

1. Additional features to help explain variation
2. For something like nerual networks you can tune depth or try using other variations on model structure (LSTM vs. GRU eg.) or for decision trees you can perform pruning
3. Hyper paramter tuning using grid search (or random search) - you can tune things like learning rate for boosted trees or penalies for logistic reg, etc.
4. If a model is overfitting, you can use ridge or lasso regression (including lamba parameters in the models)

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import cross_val_score, train_test_split
from scipy import stats
```

```python
diabetes = datasets.load_diabetes()
X = diabetes.data[:150]
y = diabetes.target[:150]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3)
```

```python
lasso = linear_model.Lasso()

glm = linear_model.LinearRegression()
```

```python
lasso_fit = lasso.fit(X_train, y_train)
```

```python
glm_fit = glm.fit(X_train,y_train)
```

```python
df_2d = pd.DataFrame(X, columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
df_1d = pd.Series(y, name='Target')
df = pd.concat([df_2d, df_1d], axis=1)
```

```python
correlation_matrix = df.corr()

# Plot correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
lasso_fit.coef_
```

```python
glm_fit.coef_
```

```python
params = np.append(glm_fit.intercept_,glm_fit.coef_)
predictions = glm_fit.predict(X_test)

newX = np.append(np.ones((len(X),1)), X, axis=1)
MSE = (sum((y_test-predictions)**2))/(len(newX)-len(newX[0]))

# Note if you don't want to use a DataFrame replace the two lines above with


var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/ sd_b

p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-len(newX[0])))) for i in ts_b]

sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

myDF3 = pd.DataFrame()
myDF3["Feature Names"] = ['intercept'] + diabetes['feature_names']
myDF3["Coefficients"],myDF3["Standard Errors"],myDF3["t values"],myDF3["Probabilities"] = [params,sd_b,ts_b,p_values]
print(myDF3)
```

```python
params = np.append(lasso_fit.intercept_,lasso_fit.coef_)
predictions = lasso_fit.predict(X_test)

newX = np.append(np.ones((len(X),1)), X, axis=1)
MSE = (sum((y_test-predictions)**2))/(len(newX)-len(newX[0]))

# Note if you don't want to use a DataFrame replace the two lines above with


var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/ sd_b

p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-len(newX[0])))) for i in ts_b]

sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

myDF3 = pd.DataFrame()
myDF3["Feature Names"] = ['intercept'] + diabetes['feature_names']

myDF3["Coefficients"],myDF3["Standard Errors"],myDF3["t values"],myDF3["Probabilities"] = [params,sd_b,ts_b,p_values]
print(myDF3)
```

```python
stats.variation(cross_val_score(lasso, X, y, cv=20))
```
